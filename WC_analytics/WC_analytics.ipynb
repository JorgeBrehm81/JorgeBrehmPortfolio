```python
from statsbombpy import sb
import pandas as pd
from mplsoccer import VerticalPitch,Pitch
from highlight_text import ax_text, fig_text
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
```

```python
free_comps = sb.competitions()
free_comps
```

```python
# Se consigue solamente los partidos del mundial del 2022
mundial_2022 = sb.matches(competition_id = 43, season_id = 106)
mundial_2022
```

```python
# Se filtra para obtener solamente los partidos de mexico
mx = mundial_2022.loc[(mundial_2022['home_team'] == 'Mexico') | (mundial_2022['away_team'] == 'Mexico')]
mx
```

```python
# para que se vean todas las columnas de la siguiente base de datos
pd.options.display.max_columns = None
```

```python
# se filtra para el juego de mx vs arg
mx_arg = sb.events(match_id = 3857289)
mx_arg
```

```python
# nombre de columnas
mx_arg.columns
```

```python
# Valores de la columna 'type'
mx_arg['type'].unique()
```

```python
# Nueva base de datos solamente de los pases del partido
pases = mx_arg[mx_arg['type'] == 'Pass']
```

```python
# coordenadas de los pases del partido
pases['location']
```

```python
# se crean columnas 'x' y 'y' para separar las coordenadas de la columna 'location'
# Se crean las columnas 'pass_end_x' y 'pass_end_y' para definir donde termina el pase
pases[['x', 'y']] = pases['location'].apply(pd.Series)
pases[['pass_end_x', 'pass_end_y']] = pases['pass_end_location'].apply(pd.Series)
pases
```

```python
# separar los pases de cada equipo
pases_mx = pases[(pases['team'] == 'Mexico')]
pases_arg = pases[(pases['team'] == 'Argentina')]
pases_mx
```

```python
# revisar nombres de los jugadores que hicieron pases de argentina
pases_mx['player'].unique()
```

```python
# recortarel nombre
pases_mx['player'] = pases_mx['player'].replace(['Héctor Miguel Herrera López'], 'Hector H')
pases_mx['player'] = pases_mx['player'].replace(['Hirving Rodrigo Lozano Bahena'], 'Chucky')
pases_mx['player'] = pases_mx['player'].replace(['Ernesto Alexis Vega Rojas'], 'Alexis V')
pases_mx['player'] = pases_mx['player'].replace(['Luis Gerardo Chávez Magallón'], 'Luis Chavez')
pases_mx['player'] = pases_mx['player'].replace(['Kevin Nahin Álvarez Campos'], 'Kevin A')
pases_mx['player'] = pases_mx['player'].replace(['Érick Gabriel Gutiérrez Galaviz'], 'Guti')
pases_mx['player'] = pases_mx['player'].replace(['Jesús Daniel Gallardo Vasconcelos'], 'Gallardo')
pases_mx['player'] = pases_mx['player'].replace(['César Jasib Montes Castro'], 'Cesar Montes')
pases_mx['player'] = pases_mx['player'].replace(['Néstor Alejandro Araújo Razo'], 'Nestor A')
pases_mx['player'] = pases_mx['player'].replace(['Francisco Guillermo Ochoa Magaña'], 'Ochoa')
pases_mx['player'] = pases_mx['player'].replace(['José Andrés Guardado Hernández'], 'Guardado')
```

```python
pases_mx['player'].unique()
```

```python
# se cuentan los pases totales por jugador de Argentina
pases_tot = pases_mx.groupby(['player'])['player'].count().to_frame()
pases_tot
```

```python
# Nan significa pase completo
pases_mx['pass_outcome'].unique()
```

```python
completos = pases_mx[pases_mx['pass_outcome'].isnull()]
incompletos = pases_mx[pases_mx['pass_outcome'].notnull()]
completos
```

```python
pases_tot['completos'] = completos.groupby(['player'])['player'].count().to_frame()
pases_tot['incompletos'] = incompletos.groupby(['player'])['player'].count().to_frame()
pases_tot = pases_tot.fillna(0)
pases_tot
```

```python
pases_tot = pases_tot.rename(columns = {'player' : 'pases totales'})
pases_tot = pases_tot.reset_index()
pases_tot['Porcentaje %'] = pases_tot['completos']/pases_tot['pases totales'] * 100
pases_tot
```

```python
pases_tot = pases_tot[pases_tot['pases totales'] > 10]
pases_tot
```

## Visualización

```python
top5_porc = pases_tot.sort_values('Porcentaje %', ascending = True).tail(5)
top5_porc
```

```python
fig = plt.figure(figsize = (14, 12))
ax = plt.axes()
ax.barh(top5_porc['player'],
       top5_porc['Porcentaje %'], color = 'green', edgecolor = 'red')
plt.title('Mejor % de pases vs Argentina')
plt.xlabel('% de pases completos')
plt.ylabel('Jugador')
plt.show
```

```python
ax = pases_tot[['player', 'completos', 'incompletos']].sort_values('completos',
                                                                  ascending = True).head(5).plot.barh(x = 'player', stacked = True)
plt.title('Pases vs Argentina')
plt.xlabel('# pases')
plt.ylabel('Jugador')
plt.show()
```

```python
white = 'white'
sbred = '#e21017'
lightgrey = '#5A5A5A'
darkgrey = '#5A5A5A'
cmaplist = [white, darkgrey, sbred]
cmap = LinearSegmentedColormap.from_list("", cmaplist)
```

```python
def passMap(player):
    player = player
    comp = completos[(completos['player'] == player)]
    incomp = incompletos[(incompletos['player'] == player)]
    
    pitch = Pitch(pitch_type = 'statsbomb', pitch_color = 'green', line_color = 'white', line_zorder = 2)
    fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('white')

    #Contando los pases por zona y coordenadas x & y de finalizacion de los pases para el mapa de calor
    '''
    bin_statistic = pitch.bin_statistic(comp.pass_end_x, comp.pass_end_y, statistic='count', bins=(12, 8), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, alpha=0.5, cmap=cmap)
    '''
    #Graficando las flechas de los pases de x jugador
    pitch.arrows(comp.x, comp.y,
    comp.pass_end_x, comp.pass_end_y, width=3,
    headwidth=8, headlength=5, color=sbred, ax=ax, zorder=2, label = "Pase completo")
    pitch.arrows(incomp.x, incomp.y,
    incomp.pass_end_x, incomp.pass_end_y, width=3,
    headwidth=8, headlength=5, color=darkgrey, ax=ax, zorder=2, label = "Pase incompleto")

    #etiquetas de color
    ax.legend(facecolor= 'white', handlelength=5, edgecolor='None', fontsize=20, loc='upper left')

    #titulo
    ax_title = ax.set_title('Pases de ' f'{player} vs Argentina', fontsize=30,color='black')
    
    plt.show()
```

```python
passMap('Guardado')
```

```python
#para crear el mismo tipo de mapa para muchos jugadores y ahorrar tiempo puedes usar una 'for loop'
#puedes utilizar cualquier base de datos donde esten los nombres de jugadores que quieres utilizar
for player in pases_tot['player'].unique():
    passMap(player)
```

```python
#Creando funcion para hacerlo de jugadores en especifico

def passMap_f3(player):
    
    player=player
    #Filtrando los pases solo del jugador seleccionado
    comp=completos[(completos["player"]==player)]
    incomp=incompletos[(incompletos["player"]==player)]
    complete_f3 = comp[(completos["pass_end_x"]>=70)]
    incomplete_f3 = incomp[(incompletos["pass_end_x"]>=70)]
    complete_f3 = complete_f3[(complete_f3["x"]<70)]
    incomplete_f3 = incomplete_f3[(incomplete_f3["x"]<70)]

    #dibujando el campo de futbol
    pitch = Pitch(pitch_type='statsbomb', pitch_color='green', line_color='white',line_zorder=2)

    fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
    fig.set_facecolor('white')

    #Contando los pases por zona y coordenadas x & y de finalizacion de los pases para el mapa de calor
    # Si quiero lo quito para quitar los cuadros rojos y grises
    '''
    bin_statistic = pitch.bin_statistic(complete_f3.pass_end_x, complete_f3.pass_end_y, statistic='count', bins=(12, 8), normalize=True)
    pitch.heatmap(bin_statistic, ax=ax, alpha=0.5, cmap=cmap)
    '''
    #Graficando las flechas de los pases de Messi
    pitch.arrows(complete_f3.x, complete_f3.y,
    complete_f3.pass_end_x, complete_f3.pass_end_y, width=3,
    headwidth=8, headlength=5, color=sbred, ax=ax, zorder=2, label = "Pase Completo")
    pitch.arrows(incomplete_f3.x, incomplete_f3.y,
    incomplete_f3.pass_end_x, incomplete_f3.pass_end_y, width=3,
    headwidth=8, headlength=5, color=darkgrey, ax=ax, zorder=2, label = "Pase Incompleto")

    #etiquetas de color
    ax.legend(facecolor='white', handlelength=5, edgecolor='None', fontsize=20, loc='upper left')

    #titulo
    ax_title = ax.set_title('Pases de ' f'{player} vs Argentina al ultimo tercio', fontsize=30,color='black')
    
    plt.show()
```
